```python
import numpy as np
import matplotlib.pyplot as plt
u = np.linspace(-2,2,65)
v = np.linspace(-1,1,33)
X,Y = np.meshgrid(u,v)
Z = X**2/25 + Y**2/4
plt.pcolor(Z) # a psuedo color plot.
#to understand the color lets diplay color bar.
plt.colorbar()
plt.show()
```

```python
#we can explcitly choose number of color maps.
#for ex
plt.pcolor(Z,cmap='gray')
#results in gray scale image
plt.colorbar()
plt.show()
```

```python
plt.pcolor(Z,cmap='autumn')
#results in yellow to orange mapping
plt.colorbar()
plt.show()
```

```python
#to fix the boundary region filling 'not in my case'
plt.pcolor(Z)
plt.colorbar()
plt.axis('tight')
plt.show()
```

In the above fig we can see that the range of the X and Y co-ordinates are different.

```python
#above is not that popular but try this
plt.pcolor(X,Y,Z)
plt.colorbar()
plt.show()
```

In the above figure we can clearly see that the range of values of the figure has changed which certainly depends upon X and Y specified

```python
#pseudo color plots tend to be blocky or pixelated.
#we can use contopur plots to the data which veries continously
plt.contour(Z)
plt.colorbar()
plt.show()
```

```python
#we can specify the number of contours to draw by speciying number in second arg
plt.contour(Z,30)
#plt.colorbar()
plt.show()
```

```python
#we can even specify the X and Y arg as in pcolor
plt.contour(X,Y,Z,30)
plt.show()
#same thing happens as the argument X and y are filled by the values in the X and Y array.
```

```python
#FILLED CONTOUR PLOTS.
plt.contourf(X,Y,Z,30)
plt.colorbar()
plt.show()
```
