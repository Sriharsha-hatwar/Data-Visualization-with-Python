Seaborn is a high level interface for drawing attractive statistical plots /graphics

Matplotlib is powerful but can be complicated.

Seaborn is  built on top of matplotlib , to make attractive statistical plots with ease.

to use seaborn we also need to know pandas 

```python
#Recapping pandas

#labelled columns and rows
#some variable are continous wheras others are categorical

#using lmplot
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
#we import matplotlib as seaborn need it.
tips = sns.load_dataset('tips')
sns.lmplot(x='total_bill' ,y='tip',data=tips) #to plot the point on the regression line,data->data frame 
#( all arguments mandatory)
plt.show()
```

```python
#our data has factors or group within our data category
#sex is categorised as male or female.
#we sometimes want to plot for different factors seperately.
#---------------------------------------------------------------------------------------
#we distinguish between male and female respectively using blue and red plots.
#they have similar slopes
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',palette='Set1')
plt.show()
#try doing for non categorical data.
#hue->imp .
#colors from color pallette.
```

```python
#we can do by not using legends but on different subplots.
sns.lmplot(x='total_bill',y='tip',data=tips,col='sex')
plt.show()
#the grid is made without using matplotlib subplot command.
```

<h2>Residual Plots</h2>

```python
#how much the data misses the rgression line.
#is plotted with residplot()
#x can be numpy arrays.
sns.residplot(x='age',y='fare',data=tips,color='indianred')
plt.show()
#optional arg color as in matplotlib.
```
