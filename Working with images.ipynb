```python
#IMAGES
#INTENSITY VALUES
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
img = plt.imread('sunflower.jpg')
```

```python
print(img.shape)
```

```python
plt.imshow(img)
```

```python
plt.show()
```

```python
plt.axis('off')
plt.imshow(img)
```

```python
#there exists number of ways to convert a colour image into a grayscale image.
#average the RGB channels by usning numpy mean method along axis 2 -> trailng dimension.
collapsed = img.mean(axis=2)

print(collapsed.shape)
```

```python
plt.set_cmap('gray')
plt.imshow(collapsed,cmap='gray')
```

```python
plt.imshow(collapsed,cmap='gray')
plt.axis('off')
plt.show()
```

```python
#one subtility regarding imshow command is the pixel shape
#suppose we take every pixel from 4th row and 2nd column ,
uneven = collapsed[::4,::2]
print(uneven.shape)
plt.imshow(uneven)
plt.axis('off')
plt.show()
```

```python
plt.imshow(uneven,aspect=2.0)
plt.axis('off')
plt.show()
```

```python
#another way wo fix this prob is to specify the original image dimensions in imshow command.
plt.imshow(uneven,cmap='grey',extent=(0,1024,0,768))
plt.show()
```
